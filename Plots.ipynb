```python
%matplotlib inline

import pandas as pd
```

```python
df_basic = pd.read_csv('independent_metrics/basic.csv', index_col='bug_id')
df_bug_inducing = pd.read_csv('independent_metrics/bug_inducing.csv', index_col='bug_id')
df = pd.merge(df_basic, df_bug_inducing, left_index=True, right_index=True)
```

## Uplifts per component

```python
df.component.value_counts().plot.pie(figsize=(8,8))
```

## Release uplifts

### Uplifts per component

```python
df[df.channels.str.contains('release')].component.value_counts().plot.pie(figsize=(8,8))
```

### Bug-inducing uplifts per component

```python
df[(df['error_inducing'] == True) & (df.channels.str.contains('release'))].component.value_counts().plot.pie(figsize=(8,8))
```

## Beta uplifts

### Uplifts per component

```python
df[df.channels.str.contains('beta')].component.value_counts().plot.pie(figsize=(8,8))
```

### Bug-inducing uplifts per component

```python
df[(df['error_inducing'] == True) & (df.channels.str.contains('beta'))].component.value_counts().plot.pie(figsize=(8,8))
```
